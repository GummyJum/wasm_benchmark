```python
import sys
from collections import defaultdict 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os.path
```

```python
total_res_dict = defaultdict(lambda: [])

def add_file(filepath, run_id):
    with open(filepath, 'r') as f:
        data = f.read().split('\n')

    res = defaultdict(lambda: [])
    for line in data:
        if len(line) < 3:
            continue

        _, name, t = line.split()
        if 'ms' in t:
            t = float(t[:-2]) / 1000
        elif 'm' in t:
            minutes, seconds = t.split('m')
            t = float(minutes) * 60 + float(seconds[:-1])
        elif 's' in t:
            t = float(t[:-1])
        res[name].append(t)
    
#     run_id = os.path.splitext(os.path.basename(filepath))[0]
    for name, r in sorted(res.items()):
        if name not in total_res_dict['Name']:
            total_res_dict['Name'].append(name)
        total_res_dict['AVG ' + run_id + ' [s]'].append(np.mean(r))
        total_res_dict['STD ' + run_id + ' [s]'].append(np.std(r))
        
add_file('native.txt', 'Go')
add_file('chrome.txt', 'Go-Wasm (chrome)')
```

```python
total_res_dict
```

```python
df = pd.DataFrame(total_res_dict)
```

```python
print(df.to_markdown())
```

```python
df.set_index('Name').plot(y=['AVG Go [s]', 'AVG Go-Wasm (chrome) [s]'], kind='bar')
plt.grid()
plt.xticks(rotation=45)
plt.ylabel('Duration [s]')
plt.xlabel('')
```

```python
plt.bar(df['Name'], df['AVG Go [s]'] / df['AVG Go-Wasm (chrome) [s]'])
plt.grid()
plt.xticks(rotation=45)
plt.ylabel('Duration Ratio [Go / Go-Wasm]')
plt.xticks(df['Name'])
```
